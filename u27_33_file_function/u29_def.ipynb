## 연습 문제
- 3자리수 정수 두개를 곱해서 나온 결과가 팔린드롬(Palindrome)이 되도록 할 때, 가장 큰 숫자는?

```python
def is_palindrome(x):
    s = str(x)
    return s == s[::-1]
```

```python
is_palindrome(12345), is_palindrome(906609)
```

```python
max_pal, m, n = 0, 0, 0
for i in range(100, 1000):
    for k in range(i, 1000):
        if is_palindrome(i*k):
            if i*k > max_pal:
                max_pal, m, n = i*k, i, k

print(f'{m} x {n} = {max_pal}')
```

# 29.3 심사문제

- 표준 입력으로 숫자 두 개가 입력됩니다. 다음 소스 코드를 완성하여 두 숫자의 덧셈, 뺄셈, 곱셈, 나눗셈의 결과가 출력되게 만드세요. 이때 나눗셈의 결과는 실수라야 합니다.

```python
x, y = map(int, input().split())

def calc(x, y):
    a = x + y
    b = x - y
    m = x * y
    d = x / y
    return a, b, m, d


a, s, m, d = calc(x, y)
print('덧셈: {0}, 뺄셈: {1}, 곱셈: {2}, 나눗셈: {3}'.format(a, s, m, d))
```
