# 27.6 심사문제

- 문자열이 저장된 words.txt 파일이 주어집니다(문자열은 한 줄로 저장되어 있습니다). words.txt 파일에서 문자 c가 포함된 단어를 각 줄에 출력하는 프로그램을 만드세요. 단어를 출력할 때는 등장한 순서대로 출력해야 하며 ,(콤마)와 .(점)은 출력하지 않아야 합니다.

```python
text = "Fortunately, however, for the reputation of Asteroid B-612, a Turkish dictator made a law that his subjects, under pain of death, should change to European costume. So in 1920 the astronomer gave his demonstration all over again, dressed with impressive style and elegance. And this time everybody accepted his report."
with open('test27.txt', 'w') as file:
    file.write(text)

with open('test27.txt') as file:
    s = file.read().split()
    for i in s:
        if i.find('c') != -1:
            if (i.find('.') != -1) | (i.find(',') != -1):
                print(i[:-1])
            else:
                print(i)

```

# 28.4 심사문제

- 단어가 줄 단위로 저장된 words.txt 파일이 주어집니다. words.txt 파일에서 회문인 단어를 각 줄에 출력하는 프로그램을 만드세요. 단어를 출력할 때는 등장한 순서대로 출력해야 합니다. 그리고 파일에서 읽은 단어는 \n이 붙어있으므로 \n을 제외한 뒤 회문인지 판단해야 하며 단어를 출력할 때도 \n이 출력되면 안 됩니다(단어 사이에 줄바꿈이 두 번 일어나면 안 됨).

```python
a = """apache\ndecal\ndid\nneep\nnoon\nrefer\nriver"""
print(a)
with open('words.txt', 'w') as file:
    file.write(a)
```

```python
with open('words.txt') as file:
    words = [x.strip('\n') for x in file.readlines()]
    for word in words:
        word1 = ''.join([word[i] for i in range(len(word))])
        word2 = ''.join([word[-1-i] for i in range(len(word))])
        if word1 == word2:
            print(word)
```

```python
with open('words.txt') as file:
    words = [x.strip('\n') for x in file.readlines()]
    for word in words:
        if word == ''.join([i for i in reversed(word)]):
            print(word)
```

# 29.3 심사문제

- 표준 입력으로 숫자 두 개가 입력됩니다. 다음 소스 코드를 완성하여 두 숫자의 덧셈, 뺄셈, 곱셈, 나눗셈의 결과가 출력되게 만드세요. 이때 나눗셈의 결과는 실수라야 합니다.

```python
x, y = map(int, input().split())

def calc(x, y):
    a = x + y
    b = x - y
    m = x * y
    d = x / y
    return a, b, m, d


a, s, m, d = calc(x, y)
print('덧셈: {0}, 뺄셈: {1}, 곱셈: {2}, 나눗셈: {3}'.format(a, s, m, d))
```

# 31.4 심사문제

- 표준 입력으로 정수 한 개가 입력됩니다(입력 값의 범위는 10~30). 다음 소스 코드를 완성하여 입력된 정수에 해당하는 피보나치 수가 출력되게 만드세요.

- 피보나치 수는 0과 1로 시작하며, 다음 번 피보나치 수는 바로 앞의 두 피보나치 수의 합입니다.

```python
def fib(n):
    _a, _b = 0, 1
    for _ in range(n):
        _a, _b = _b, _a + _b
    return _a



n = int(input())
print(fib(n))
```
