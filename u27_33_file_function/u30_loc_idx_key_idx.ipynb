## 30.1 loc

```python
def print_numbers(a,b,c):
    print(a)
    print(b)
    print(c)
```

```python
print_numbers(10,20,30)
```

```python
# *x, *y - 리스트/튜플 언패킹
x, y = [10,20,30], (40,50,60)
print_numbers(*x)
print_numbers(*y)
```

### 가변 인수 함수

```python
def print_numbers2(*args):
    for arg in args:
        print(arg)
```

```python
print_numbers2(10,20,30,40)
```

```python
print_numbers2(10)
```

### 고정 인수와 가변 인수

```python
def print_numbers3(a, *args):
    print(a)
    print(args)
```

```python
print_numbers3(1, 10, 20)
```

```python
print_numbers3(*x)
```

```python
# 고정 위치 인수가 가변 위치 인수보다 먼저 나와야 함!!!
def print_numbers4(*args, a):
    print(args)
    print(a)
```

```python
print_numbers4(*x)
```

## 30.2 키워드 인수

```python
def personal_info(name, age, addr):
    print('이름:', name)
    print('나이:', age)
    print('주소:', addr)
```

```python
personal_info('홍길동',30,'서울 강남구 도곡동')
```

```python
personal_info(age=30,addr='서울 강남구 도곡동',name='홍길동')
```

## 30.3 키워드 인수와 딕셔너리 언패킹

```python
x = {'name':'홍길동', 'age':30, 'addr':'서울 강남구 도곡동'}
personal_info(**x)
```

```python
# *x는 x의 키를 불러냄
personal_info(*x)
```

### 가변 키워드 인수

```python
def personal_info2(**kwargs):
    for key, value in kwargs.items():
        print(f'{key}: {value}')
```

```python
personal_info2(**x)
```

```python
personal_info2(name='홍길동',age=30)
```

### 고정 인수와 가변 키워드 인수

```python
def personal_info3(name, **kwargs):
    print(name)
    print(kwargs)
```

```python
personal_info3(**x)
```

### 위치 인수와 키워드 인수

```python
def my_print(*args, **kwargs):
    print(*args, **kwargs)
```

```python
my_print(10,20,30, sep=', ')
```

# 30.6 심사문제

- 표준 입력으로 국어, 영어, 수학, 과학 점수가 입력됩니다. 다음 소스 코드를 완성하여 가장 높은 점수, 가장 낮은 점수, 평균 점수가 출력되게 만드세요. 평균 점수는 실수로 출력되어야 합니다.

```python
korean, english, mathematics, science = map(int, input().split())

def get_min_max_score(*args):
    return min(args), max(args)
def get_average(**kwargs):
    return (sum(kwargs.values())/len(kwargs))

min_score, max_score = get_min_max_score(korean, english, mathematics, science)


average_score = get_average(korean=korean, english=english,
                            mathematics=mathematics, science=science)
print('낮은 점수: {0:.2f}, 높은 점수: {1:.2f}, 평균 점수: {2:.2f}'
      .format(min_score, max_score, average_score))
 
min_score, max_score = get_min_max_score(english, science)
average_score = get_average(english=english, science=science)
print('낮은 점수: {0:.2f}, 높은 점수: {1:.2f}, 평균 점수: {2:.2f}'
      .format(min_score, max_score, average_score))
```
