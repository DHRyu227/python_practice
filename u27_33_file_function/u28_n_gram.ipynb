```python
# N-gram
# list에 * 을 붙이는 것은 언패킹(list unpacking) 이라함.
text = 'hello'
[text[i:] for i in range(3)]
```

```python
# 연습문제

n = int(input())
text = input()
words = text.split()    
 
if (len(words)) < n:
    print('wrong')
else:
    n_gram = zip(*[words[i:] for i in range(n)])                              
    for i in n_gram:
        print(i)
```

# 28.4 심사문제

- 단어가 줄 단위로 저장된 words.txt 파일이 주어집니다. words.txt 파일에서 회문인 단어를 각 줄에 출력하는 프로그램을 만드세요. 단어를 출력할 때는 등장한 순서대로 출력해야 합니다. 그리고 파일에서 읽은 단어는 \n이 붙어있으므로 \n을 제외한 뒤 회문인지 판단해야 하며 단어를 출력할 때도 \n이 출력되면 안 됩니다(단어 사이에 줄바꿈이 두 번 일어나면 안 됨).

```python
a = """apache\ndecal\ndid\nneep\nnoon\nrefer\nriver"""
print(a)
with open('words.txt', 'w') as file:
    file.write(a)
```

```python
with open('words.txt') as file:
    words = [x.strip('\n') for x in file.readlines()]
    for word in words:
        word1 = ''.join([word[i] for i in range(len(word))])
        word2 = ''.join([word[-1-i] for i in range(len(word))])
        if word1 == word2:
            print(word)
```

```python
with open('words.txt') as file:
    words = [x.strip('\n') for x in file.readlines()]
    for word in words:
        if word == ''.join([i for i in reversed(word)]):
            print(word)
```
